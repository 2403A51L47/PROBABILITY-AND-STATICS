```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

mu = 20
sigma = 4
x1 = 23
x2 = 35

z1 = (x1 - mu) / sigma
z2 = (x2 - mu) / sigma

prob_ge_23 = 1 - stats.norm.cdf(z1)
prob_le_23 = stats.norm.cdf(z1)
prob_between_23_and_35 = stats.norm.cdf(z2) - stats.norm.cdf(z1)

prob_ge_23, prob_le_23, prob_between_23_and_35

x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Normal Distribution (BMI)', color='blue')

plt.fill_between(x_vals, y_vals, where=(x_vals >= x1), color='red', alpha=0.5, label="P(X ≥ 23)")
plt.fill_between(x_vals, y_vals, where=(x_vals <= x1), color='green', alpha=0.5, label="P(X ≤ 23)")
plt.fill_between(x_vals, y_vals, where=(x_vals >= x1) & (x_vals <= x2), color='yellow', alpha=0.5, label="P(23 ≤ X ≤ 35)")

plt.title('BMI Normal Distribution with Highlighted Probabilities')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

mean = 75
std_dev = 10
num_points = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=num_points)

plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='skyblue', stat='density')

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, loc=mean, scale=std_dev)
plt.plot(x, pdf, color='red', linewidth=2, label="PDF")

plt.title('Histogram and Probability Density Function (PDF) of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.show()
plt.figure(figsize=(10, 6))
```
