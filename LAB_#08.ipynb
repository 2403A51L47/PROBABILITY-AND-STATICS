```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores, color='blue')
plt.title('Study Hours vs. Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
```

```python
plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores, color='green')
plt.title('Sleep Hours vs. Exam Scores')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Scores')
```

```python

plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours, color='red')
plt.title('Study Hours vs. Sleep Hours')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.tight_layout()
plt.show()
```

```python
r_study_exam, _ = pearsonr(study_hours, exam_scores)
r_sleep_exam, _ = pearsonr(sleep_hours, exam_scores)
r_study_sleep, _ = pearsonr(study_hours, sleep_hours)
```

```python
print(f"Pearson correlation between Study Hours and Exam Scores: {r_study_exam:.2f}")
print(f"Pearson correlation between Sleep Hours and Exam Scores: {r_sleep_exam:.2f}")
print(f"Pearson correlation between Study Hours and Sleep Hours: {r_study_sleep:.2f}")
```

```python
if r_study_exam > 0.7:
    print("There is a strong positive correlation between Study Hours and Exam Scores.")
elif r_study_exam > 0.3:
    print("There is a moderate positive correlation between Study Hours and Exam Scores.")
else:
    print("There is a weak or no correlation between Study Hours and Exam Scores.")

if r_sleep_exam > 0.7:
    print("There is a strong positive correlation between Sleep Hours and Exam Scores.")
elif r_sleep_exam > 0.3:
    print("There is a moderate positive correlation between Sleep Hours and Exam Scores.")
else:
    print("There is a weak or no correlation between Sleep Hours and Exam Scores.")

if r_study_sleep > 0.7:
    print("There is a strong positive correlation between Study Hours and Sleep Hours.")
elif r_study_sleep > 0.3:
    print("There is a moderate positive correlation between Study Hours and Sleep Hours.")
else:
    print("There is a weak or no correlation between Study Hours and Sleep Hours.")
```
