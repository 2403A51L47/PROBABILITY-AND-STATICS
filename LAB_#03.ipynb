```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('employee.csv')
df.info()
```

```python
df.head()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.histplot(df["Salary"], bins=20, kde=True, color="skyblue")
plt.xlabel("Salary (USD)")
plt.ylabel("Frequency")
plt.title("Salary Distribution of Employees")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Experience (Years)"], y=df["Salary"], hue=df["Job Satisfaction"], palette="coolwarm", alpha=0.7)
sns.regplot(x=df["Experience (Years)"], y=df["Salary"], scatter=False, color="black", line_kws={"linewidth": 1})
plt.xlabel("Years of Experience")
plt.ylabel("Salary (USD)")
plt.title("Experience vs Salary")
plt.legend(title="Job Satisfaction")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Department"], palette="viridis", order=df["Department"].value_counts().index)
plt.xlabel("Number of Employees")
plt.ylabel("Department")
plt.title("Number of Employees in Each Department")
plt.show()
```
